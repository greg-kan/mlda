```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, np.pi * 3, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
fig.savefig('my_figure.png')
```
