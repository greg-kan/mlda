```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
%matplotlib inline
```

```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
# Используем серый фон
ax = plt.axes(facecolor='#E6E6E6')
ax.set_axisbelow(True)
# Рисуем сплошные белые линии сетки
plt.grid(color='w', linestyle='solid')
# Скрываем основные линии осей координат
for spine in ax.spines.values():
    spine.set_visible(False)
# Скрываем деления сверху и справа
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
# Осветляем цвет делений и меток
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
# Задаем цвет заливки и границ гистограммы
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666')
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
# Восстанавливаем rcParams
plt.rcParams.update(IPython_default);
```

```python
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```
