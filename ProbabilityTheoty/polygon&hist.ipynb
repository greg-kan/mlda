```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 100000
```

```python
y = np.random.randn(N)
```

```python
plt.hist(y, 25)
plt.show()
```

```python
len(y)
```

```python
y[y > 4]
```

```python
def histogram(data, n_bins = 4, cumulative=False, density = False, x_label = "", y_label = "", title = ""):
    _, ax = plt.subplots()
    ax.hist(data, bins = n_bins, cumulative = cumulative, density = density, color = '#539caf')
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
```

```python
histogram(y)
```

```python
histogram(y, 25, False, "X", "n", "Test Histogram")
```
