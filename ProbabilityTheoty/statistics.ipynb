```python
from numpy import random
r = random.random()            # одно число между 0 и 1
r = random.random(size=10000)  # массив с 10000 чисел
r = random.uniform(-1, 10)     # одно число между -1 и 10
r = random.uniform(-1, 10, size=10000)   # массив
```

```python
N = 10**6
from numpy import random, mean, var, std
x = random.uniform(-1, 1, size=N)
xm = mean(x)  # среднее значение
xv = var(x)   # дисперсия
xs = std(x)   # СКО
print('''Number: %d
mean:   %.2e
var:    %.2e
stdev:  %.2e''' % (N, xm, xv, xs))
```

# Генерация массива из словаря

```python
voc = {4:10, 7:15, 10:20, 15:5}
```

```python
print(voc[7])
```

```python
for key in voc:
    print("%s -> %s" % (key, voc[key]))
```

```python
z = np.array([])
```

```python
for key in voc:
    for i in range(voc[key]):
        #print("%s -> %s" % (key, voc[key]))
        z = np.append(z, key)
        
print(z)
```

```python
num = len(z)
zm = mean(z)  # среднее значение
zv = var(z)   # дисперсия
zs = std(z)   # СКО
print('''Number: %d
mean:   %.2e
var:    %.2e
stdev:  %.2e''' % (N, zm, zv, zs))
```
