# Аппроксимация функций

### f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2) on x in [1, 15] 

```python
import numpy as np
```

```python
from matplotlib import pylab as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
init_values = [1, 4, 10,  15]
```

```python
iv_len = len(init_values)
```

```python
matrix = np.ones((iv_len, iv_len))
```

```python
for i in range(iv_len):
    for j in range(iv_len):
        matrix[i,j] = init_values[i]**j
        
print(matrix)
```

```python
free_members = np.ones((iv_len))
for i in range(iv_len):
    free_members[i] = f(init_values[i])
```

```python
from scipy import linalg
```

```python
w = np.linalg.solve(matrix, free_members)
print(w)
```

```python
def f_aprox(x):
    res = 0.0
    for i in range(iv_len):
        res += w[i]*(x**i)
    return res
```

```python
x = np.arange(0.5, 15.5, 0.1)
plt.plot(x, f(x))
plt.show()
```

```python
plt.plot(x, f_aprox(x))
plt.show()
```
