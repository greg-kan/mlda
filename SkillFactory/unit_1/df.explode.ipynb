```python
def split_index(data):
    """
    Функция принимает в качестве параметра объект pandas series
    со строковым индексом, содержащим разделители '|',
    и числовой колонкой.
    Возвращает объкет series с индексом, состоящем из
    всех элементов индекса параметра, полученых при
    помощи split('|'), и числовой колонкой, содержащей
    суммы чисел из соответствующих строк параметра.
    """
    data_new = pd.Series()
    for ind in data.index:
        index_list = ind.split('|')
        for index_item in index_list:
            if (index_item in data_new.index):
                data_new[index_item] += data[ind]
            else:
                data_new[index_item] = data[ind]
    return data_new
```

```python
"""
Данные можно и не групировать, все зависит от того, насколько размер датасета уменьшается в результате группировки
В следующем вопросе с актерами думаю точно стоит.
"""

split_index(data.groupby(['director'])['profit'].sum()).sort_values(ascending=False).head()
```

```python
dt = data[data.production_companies.str.contains('Warner')].production_companies.str.split('|')
   
def split_test(data):
    '''
    Формирует на выходе series с индексом из компаний, содержащих в названии "Warner"
    с целью посмотреть, какие компании имеют отношение к "Warner"
    '''
    data_new = pd.Series()
    for ind in data.index:
        for item in data[ind]:
            if 'Warner' in item:
                if item in data_new.index:
                    data_new[item] += 1
                else:
                    data_new[item] = 1
    return data_new

split_test(dt)
```

# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
data_new = pd.Series()
def convert_f1(row):
    '''
    Функция записывает в серию data_new названия компаний в качестве индекса и
    tuple вида (средняя длина, длина, количество)
    '''
    for ind in row['production_companies'].split('|'):
        if (ind in data_new.index):
            mid, len, count = data_new[ind]
            len += row['title_len']
            count += 1
            mid = len / count
            data_new[ind] = (mid, len, count)
        else:
            data_new[ind] = (row['title_len'], row['title_len'], 1)  

data.apply(convert_f1, axis=1)
data_new.head()
```

```python
#Затем выбираем среднюю длину и сортируем
data_new.apply(lambda x: x[0]).sort_values(ascending=False)[:5]
```
