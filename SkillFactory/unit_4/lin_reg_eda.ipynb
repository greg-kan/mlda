```python
data.dropna(how='any', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=77)
```

```python
#Scaling 
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler()
X_train_scal = transformer.fit_transform(X_train)
X_test_scal = transformer.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
myModel = LinearRegression()
myModel.fit(X_train_scal, y_train)
```

```python
y_pred = myModel.predict(X_test_scal)
y_pred
```

```python
np.exp(y_pred)
```

```python
np.exp(y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE = mean_squared_error(y_pred, y_test)
print(MSE)
```

```python
mean_squared_error(np.exp(y_test) - 1, np.exp(y_pred) - 1)
```
