```python
import numpy as np
import matplotlib.pyplot as plt
import random
import statsmodels.api as sm
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
num = 100
```

```python
eps = np.random.normal(0, 1, size=num)
```

```python
X1 = np.zeros(num)
```

```python
X1[0] = eps[0]
```

```python
X1[1] = eps[1]
```

```python
for i in range (2, num):
    X1[i] = 0.2*X1[i-1] + 0.1*X1[i-2] + eps[i]
```

```python
plt.plot(X1)
```

```python
test = sm.tsa.adfuller(X1)
print ('p-value: ', test[1])
if test[1]< 0.05: 
    print (' ряд стационарен')
else:
    print ('ряд не стационарен')
```

```python
for i in range (2, num):
    X1[i] = 0.9*X1[i-1] + 0.07*X1[i-2] + eps[i]
```

```python
plt.plot(X1)
```

```python
test = sm.tsa.adfuller(X1)
print ('p-value: ', test[1])
if test[1]< 0.05: 
    print (' ряд стационарен')
else:
    print ('ряд не стационарен')
```
