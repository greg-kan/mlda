Задание: Написать код, который запрашивает данные о температуре комфорта в Москве и любом другом городе (по желанию автора) и сравнивает их

```python
def temp(href):
    r = requests.get(href)
    soup = BeautifulSoup(r.text, 'html.parser')
    temp = soup.find(class_ = 'h5 feeled-temperature').find(class_ = 'value').text[:-1]
    temp = int(temp)
    return temp
temp_MSK = temp("http://www.meteoservice.ru/weather/now/moskva.html")
temp_SPB = temp("https://www.meteoservice.ru/weather/now/sankt-peterburg-primorskiy")
if temp_MSK < temp_SPB:
    print("В питере теплее")
else:
    print("Как обычно")
```

```python
tables = pd.read_html('https://www.meteoservice.ru/weather/today/moskva')
df = tables[0]
df.dropna(inplace = True)
df.drop(['Погода', 'Вероятность осадков', 'Вероянт. осадков', 'Ветер', 'Ветер.1', 'Давление', 'Влажность'], 
        axis = 1, inplace = True)
df['Время'] = df['Время'].apply(lambda x: int(x[:-3]))

def spliter(i):
    i = i.split("/")
    Real_temp = int(i[0].strip()[:-1])
    Feal_temp = int(i[1].strip()[:-1])
    return Real_temp, Feal_temp

df["Real_temp"], df["Feal_temp"] = zip(*df["Температура/ Ощущается"].apply(spliter))
df.drop('Температура/ Ощущается', axis = 1, inplace = True)
```
